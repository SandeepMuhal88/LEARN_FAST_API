```python
data=df.copy()
#1.BMi feature
data['bmi']=data['weight']/(data['height']**2)

#2.age group feature
def age_group(age):
    if age < 25:
        return "young"
    elif age < 45:
        return "adult"
    elif age < 60:
        return "middle_aged"
    return "senior"
data['age_group']=data['age'].apply(age_group)

```

```python
# Feature 3: Lifestyle Risk
def lifestyle_risk(row):
    if row["smoker"] and row["bmi"] > 30:
        return "high"
    elif row["smoker"] or row["bmi"] > 27:
        return "medium"
    else:
        return "low"
```
